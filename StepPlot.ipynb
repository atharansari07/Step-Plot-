```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
csv=pd.read_csv("D:\\PythonProgramme\\Panda\\Employee.csv")
csv
```

```python
x=csv["Name"]
x
```

```python
y=csv["TotalSalary"]
y
```

```python
plt.figure(figsize=(18,5))
plt.step(x,y,marker="o",color="g",ms=8,mfc="b",label="Employee's Salary")
plt.xlabel("Employee's Name",fontsize=15)
plt.ylabel("Employee's Salary",fontsize=15)
plt.title("Employee Salary Details",fontsize=15)
plt.legend()
plt.grid()
plt.show()
```
